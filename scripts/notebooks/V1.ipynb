```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from io import BytesIO

# === Load Excel file ===
file_path = "CPI_DATA.xlsx"
xls = pd.ExcelFile(file_path)

# Load sheets
singapore_df = xls.parse('Singapore_CPI')
switzerland_df = xls.parse('Switzerland_CPI')

# === Step 1: Annual Data ===
annual_singapore = singapore_df[["Year", "Annual"]].dropna()
annual_switzerland = switzerland_df[["Year", "Annual"]].dropna()
annual_combined = pd.merge(
    annual_singapore, annual_switzerland, on="Year", how="outer",
    suffixes=(" Singapore", " Switzerland")
)

# === Step 2: Reshape Monthly Data ===
def reshape_cpi_data(df, country_name):
    df = df.drop(columns=["Annual"], errors='ignore')
    df_melted = df.melt(id_vars="Year", var_name="Month", value_name=country_name)
    df_melted["Month"] = df_melted["Month"].str.capitalize()
    df_melted["Date"] = pd.to_datetime(df_melted["Year"].astype(str) + "-" + df_melted["Month"], format="%Y-%b")
    df_final = df_melted[["Date", country_name]].sort_values("Date").reset_index(drop=True)
    return df_final

singapore_cpi = reshape_cpi_data(singapore_df, "Singapore")
switzerland_cpi = reshape_cpi_data(switzerland_df, "Switzerland")



# Merge monthly data
combined_df = pd.merge(singapore_cpi, switzerland_cpi, on="Date", how="outer").sort_values("Date").reset_index(drop=True)
combined_df["Year/Month"] = combined_df["Date"].dt.to_period("M").astype(str)
monthly_data = combined_df[["Year/Month", "Singapore", "Switzerland"]]

# === Step 3: EDA ===
eda_summary = monthly_data[["Singapore", "Switzerland"]].describe().T
correlation_matrix = monthly_data[["Singapore", "Switzerland"]].corr().T

# === Step 4: Add Dummy Economic Indicators and Log Transform ===
monthly_data['CPI_Singapore'] = monthly_data['Singapore']
monthly_data['CPI_Switzerland'] = monthly_data['Switzerland']
monthly_data['Nominal'] = np.random.uniform(1000, 5000, len(monthly_data))
monthly_data['Real'] = monthly_data['Nominal'] / (1 + (monthly_data['CPI_Singapore'].fillna(method='ffill') / 100))

monthly_data.dropna(inplace=True)

# Apply natural log
columns_to_log = ["CPI_Singapore", "CPI_Switzerland", "Nominal", "Real"]
for col in columns_to_log:
    monthly_data[f"log_{col}"] = np.log(monthly_data[col])

# === Step 5: Create Visualizations ===
sns.set(style="whitegrid")
plot_buffers = []

# Plot 1: CPI Trend
buf1 = BytesIO()
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(monthly_data['Year/Month']), monthly_data['Singapore'], label='Singapore', color='orange')
plt.plot(pd.to_datetime(monthly_data['Year/Month']), monthly_data['Switzerland'], label='Switzerland', color='gold')
plt.title("CPI Trend Over Time")
plt.xlabel("Date")
plt.ylabel("CPI")
plt.legend()
plt.tight_layout()
plt.savefig(buf1, format='png')
buf1.seek(0)
plot_buffers.append(('CPI Trend Over Time', buf1))

# Plot 2: Histogram
buf2 = BytesIO()
plt.figure(figsize=(10, 6))
sns.histplot(monthly_data['Singapore'], bins=30, kde=True, color='royalblue', label='Singapore', alpha=0.6)
sns.histplot(monthly_data['Switzerland'], bins=30, kde=True, color='seagreen', label='Switzerland', alpha=0.6)
plt.title("Histogram of CPI Values")
plt.xlabel("CPI")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.savefig(buf2, format='png')
buf2.seek(0)
plot_buffers.append(('Histogram of CPI Values', buf2))

# Plot 3: Correlation Heatmap
buf3 = BytesIO()
plt.figure(figsize=(6, 4))
sns.heatmap(monthly_data[["Singapore", "Switzerland"]].corr(), annot=True, cmap='RdBu_r', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig(buf3, format='png')
buf3.seek(0)
plot_buffers.append(('Correlation Heatmap', buf3))

# === Step 6: Save Final Excel File ===
output_path = "CPI_Processed_Data_with_Plots_v1.xlsx"
with pd.ExcelWriter(output_path, engine='xlsxwriter') as writer:
    monthly_data.to_excel(writer, sheet_name="Monthly Data with Logs", index=False)
    annual_combined.to_excel(writer, sheet_name="Annual Data", index=False)
    eda_summary.to_excel(writer, sheet_name="EDA Summary")
    correlation_matrix.to_excel(writer, sheet_name="Correlation Matrix")

    workbook = writer.book
    for title, buffer in plot_buffers:
        worksheet = workbook.add_worksheet(title[:30])
        worksheet.insert_image('B2', title + ".png", {'image_data': buffer})
```

**ADF Test – Stationarity Check**

```python
def adf_test(series, name):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Test for {name}')
    print(f'  Test Statistic   : {result[0]:.4f}')
    print(f'  p-value          : {result[1]:.4f}')
    print(f'  # Lags Used      : {result[2]}')
    print(f'  # Observations   : {result[3]}')
    print('  Critical Values  :')
    for key, value in result[4].items():
        print(f'    {key}: {value:.4f}')
    print('  Result: ', 'Stationary (I(0))' if result[1] < 0.05 else 'Non-stationary')
    print('-' * 50)
```

```python
# Running ADF test
adf_test(df['log_neer'], 'Log Nominal Exchange Rate')
adf_test(df['log_reer'], 'Log Real Exchange Rate')
adf_test(df['log_cpi_home'], 'Log CPI - Home Country')
adf_test(df['log_cpi_foreign'], 'Log CPI - Foreign Country')
```

```python
# Create first differences as they are all Non-stationary
df['dlog_neer'] = df['log_neer'].diff()
df['dlog_reer'] = df['log_reer'].diff()
df['dlog_cpi_home'] = df['log_cpi_home'].diff()
df['dlog_cpi_foreign'] = df['log_cpi_foreign'].diff()

# Drop NA from differencing
df_diff = df.dropna()

# ADF on first-differenced series
adf_test(df_diff['dlog_neer'], 'Diff Log Nominal Exchange Rate')
adf_test(df_diff['dlog_reer'], 'Diff Log Real Exchange Rate')
adf_test(df_diff['dlog_cpi_home'], 'Diff Log CPI - Home')
adf_test(df_diff['dlog_cpi_foreign'], 'Diff Log CPI - Foreign')
```

# Testing Absolute form of Purchasing Power Parity

```python
# Step 1: Create CPI differential
df['cpi_diff'] = df['log_cpi_home'] - df['log_cpi_foreign']

# Step 2: Run regression: log_neer ~ cpi_diff
X_abs = sm.add_constant(df['cpi_diff'])
y_abs = df['log_neer']
model_abs_ppp = sm.OLS(y_abs, X_abs).fit()

# Print regression summary
print(model_abs_ppp.summary())

# Step 3: Extract residuals and run ADF test
residuals_abs = model_abs_ppp.resid

# ADF Test Function
def adf_test(series, title=''):
    print(f"ADF Test for {title}")
    result = adfuller(series.dropna())
    print(f"  Test Statistic   : {result[0]:.4f}")
    print(f"  p-value          : {result[1]:.4f}")
    print(f"  # Lags Used      : {result[2]}")
    print(f"  # Observations   : {result[3]}")
    for key, val in result[4].items():
        print(f"    {key}%: {val:.4f}")
    if result[1] <= 0.05:
        print("Result: Stationary (reject H0)")
    else:
        print("Result: Non-stationary (fail to reject H0)")
    print("-" * 50)

# Run ADF on residuals
adf_test(residuals_abs, 'Residuals from Absolute PPP Regression')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['cpi_diff'], df['log_neer'], alpha=0.6, label='Data')
plt.plot(df['cpi_diff'], model_abs_ppp.predict(X_abs), color='red', label='Fitted Line')
plt.title("Absolute PPP: Log Exchange Rate vs CPI Differential")
plt.xlabel("CPI Differential (Home - Foreign)")
plt.ylabel("Log Nominal Exchange Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# ARIMA Model Comparison Table

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
df_clean = df.dropna()

orders = [(1, 1, 0), (0, 1, 1), (1, 1, 1), (2, 1, 0), (0, 1, 2), (2, 1, 2)]
comparison_dict = {}

for order in orders:
    model_name = f"ARIMA{order}"
    try:
        model = ARIMA(df_clean['log_reer'], order=order)
        result = model.fit()
        resid = result.resid

        # Ljung-Box Q-test
        q_stats = acorr_ljungbox(resid, lags=[4, 8, 12], return_df=True)

        # Coefficients and t-stats
        params = result.params
        tstats = result.tvalues

        stats_dict = {
            "a0": f"{params[0]:.4f} ({tstats[0]:.4f})" if len(params) > 0 else "",
            "a1": f"{params[1]:.4f} ({tstats[1]:.4f})" if len(params) > 1 else "",
            "a2": f"{params[2]:.4f} ({tstats[2]:.4f})" if len(params) > 2 else "",
            "b1": f"{params[3]:.4f} ({tstats[3]:.4f})" if len(params) > 3 else "",
            "b2": f"{params[4]:.4f} ({tstats[4]:.4f})" if len(params) > 4 else "",
            "AIC": round(result.aic, 4),
            "BIC": round(result.bic, 4),
            "Q(4)": round(q_stats['lb_stat'].iloc[0], 4),
            "Q(4) p": round(q_stats['lb_pvalue'].iloc[0], 4),
            "Q(8)": round(q_stats['lb_stat'].iloc[1], 4),
            "Q(8) p": round(q_stats['lb_pvalue'].iloc[1], 4),
            "Q(12)": round(q_stats['lb_stat'].iloc[2], 4),
            "Q(12) p": round(q_stats['lb_pvalue'].iloc[2], 4),
        }

        # Add to master comparison dict
        comparison_dict[model_name] = stats_dict

    except Exception as e:
        print(f"{model_name} failed: {e}")

# --- STEP 3: Format into Transposed Table ---
comparison_df = pd.DataFrame(comparison_dict).reset_index()
comparison_df.rename(columns={"index": "Term"}, inplace=True)

# --- STEP 4: Display Side-by-Side in Notebook ---
from IPython.display import display
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("\nARIMA Model Comparison Table (Transposed):")
display(comparison_df)

# --- STEP 5: Save to Excel ---
comparison_df.to_excel("ARIMA_Model_Comparison_Transposed.xlsx", index=False)
```
